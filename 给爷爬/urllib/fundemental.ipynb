```python
from urllib import request,parse
from pprint import pprint
```

```python
data = bytes(parse.urlencode({'world': 'hello'}),encoding='utf-8')
response = request.urlopen("http://httpbin.org/post",data=data)
pprint(response.read)
response = request.urlopen('http://httpbin.org/post', timeout=1)
# 超过时间就会报错。
```

如果网站时间过长不反应我们可以跳过它

```python
import socket
import urllib.error
```

```python
# 跳过超时的url
try:
    response = urllib.request.urlopen('http://httpbin.org/get',timeout=0.25)
except urllib.error.URLError as e:
    if isinstance(e.reason, socket.timeout):
        print('Time out!')
```

class urllib.request.Request(url, data=None, headers={}, origin_req_host=None, unverifiable=False, method =None)
* 可以发现我们的爬取的对象可以建立一个对象来使用，需要时候随时调用随时爬取。
> 参数
* 第一个参数是： URL这个是必须要传入的参数，其他的为选择性参数
* 第二个参数是： data 如果要传入这个参数的话，必须要是字节流类型的，如果它是字典，可以先用urllib.parse模块里面的urlencode()编码
* 第三个参数是: headers这个参数是放置请求头的地方，我们在构造的适合请求通过headers参数直接构造，也可以通过add_header()方法来后续添加
> 添加请求头最常用的手法就是通过修改User-Agent 来伪装浏览器，默认的User-Agent是Python-urllib,我们可以修改它来伪装浏览器，比如伪装为Edge浏览器为
