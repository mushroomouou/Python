# Ajax 数据爬取

* 再网页打开过后并不是一下子全部加载出来，而是再向服务端发送请求，请求某个接口，再拿着从服务端获得的数据加载网页，这就是Ajax加载方法

* 按照网页的发展趋势，后面的网页将更多的采用Ajax请求方式，由返回的数据逐步渲染网页，降低了服务器一次性渲染的压力，实现前后端分离

## 1.什么是Ajax
> Asynchronous JavaScript and XML,是利用JavaScrip在保证页面不被刷新， 链接不被改变的前提下，与服务端交换数据，渲染改变页面的技术

## 2.基本原理
* 发送Ajax请求
* 解析内容
* 渲染网页
